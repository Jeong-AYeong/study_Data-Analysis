```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

##### 선그래프

```python
#그래프 제목 설정
plt.title("title")
#그래프 축 이름 설정
plt.xlabel("x")
plt.ylabel("y")
#그래프 x축, y축 값, 그래프 스타일 지정
plt.plot([10, 20, 30, 40], [1, 10, 100, 1000], 'rs--')
plt.show()
```

#####  세로막대그래프

```python
sales = [25, 35, 15]    #y축 값
menu = ['americano','cafelatte','macciyatto']   #x축값
plt.title("Bar Chart")
plt.xlabel("menu")
plt.ylabel("count")
#x축은 메뉴, y축은 판매수
plt.bar(menu, sales)
plt.show()
```

##### 수평 막대그래프

```python
sales = [25, 35, 15]    #y축 값
menu = ['americano','cafelatte','macciyatto']   #x축값
plt.title("Bar Chart")
plt.xlabel("menu")
plt.ylabel("count")
#수평막대그래프 : 인자 순서주의!!!!
plt.barh(menu, sales)
plt.show()
```

##### 원그래프

##### 여러개의 그래프 그리기

```python
plt.subplot(2,2,1)
ratio=[34, 32, 16, 18]

labels=['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0.05, 0.05, 0.05, 0.05]

colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
plt.pie(ratio, labels=labels, colors=colors)
plt.title("data 1")

plt.subplot(2,2,2)
sales=[25, 35, 15]
menu=['americano','cafelatte','macciyatto']
plt.barh(menu, sales)
plt.title("data 2")

plt.subplot(2,2,3)
x= np.random.randn(1000)
plt.hist(x, bins=100)
plt.title("data 3")

plt.subplot(2,2,4)

data1= np.random.normal(0, 2.0, 1000)
data2= np.random.normal(-3.0, 1.5, 1000)
data3= np.random.normal(1.2, 1.5, 1000)
plt.boxplot( [data1, data2, data3])
plt.title("data 4")
plt.tight_layout()
plt.show()
```
