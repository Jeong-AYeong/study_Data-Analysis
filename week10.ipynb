#### 커널밀도 추정 => 히스토그램 분포를 부드럽게 선으로 표현

```python
df2 = df.drop("caseno", axis = 1)
df2.plot.kde()
plt.show()
```

```python
#groupby를 이용하여 iris 데이터셋에서 꽃의 종류에 따른 특징값 평균 구하기
df2 = df.drop("caseno", axis = 1)
gb = df2.groupby("Species").mean()
print(gb)
print()
print(gb.T)
print()
gb.plot.bar(rot = 0)
plt.title("Feature  mean of each species")
plt.ylabel("mean")
plt.show()
gb.T.plot.bar(rot = 0)
plt.title("Feature  mean of each species")
plt.ylabel("mean")
plt.show()
```

## seaborn 라이브러리

```python
import seaborn as sns
iris = sns.load_dataset("iris")
print(iris)
```

#### scatterplot

```python
#1)일반 scatterplot
#sns.scatterplot(data = iris, x = "petal_length", y = "petal_width")
#plt.show()

#2)hue인자 사용한 scatterplot
sns.scatterplot(data = iris, x = "petal_length", y = "petal_width", hue = "species")
plt.show()
```

#### histplot

```python
#1)일반 histplot
#sns.histplot(data = iris, x = "petal_length", bins = 10)
#plt.show()

#2)hue인자 사용한 histplot
sns.histplot(data = iris, x = "petal_length", bins = 10, hue = "species")
plt.show()
```

#### boxplot

```python
#boxplot
plt.subplot(2,2,1)
sns.boxplot(data = iris)

plt.subplot(2,2,2)
sns.boxplot(data = iris, x = "species", y = "petal_width")
plt.show()
```

#### jointplot

```python
#1)sns.jointplot(data = iris, x = "sepal_length", y = "sepal_width")
#plt.show()

#2)hue인자 사용한 jointplot
sns.jointplot(data = iris, x = "sepal_length", y = "sepal_width", hue = "species")
plt.show()
```

#### pairplot : 각 데이터 열 조합에 대해 scatter plot/같은 데이터 만나면 히스토그램

```python
sns.pairplot(data = iris, hue = "species")
```

## 시각화 연습 flight

```python
flights = sns.load_dataset("flights")
print(flights)
print(flights['year'].value_counts())  #결측치 확인용
```

```python
#연도별 승객 수 시각화
sns.lineplot(data = flights, x = "year", y = "passengers")
plt.show()
sns.lineplot(data = flights, x = "year", y = "passengers", hue = "month")
plt.show()
```

```python
#월별 승객 수 시각화
sns.lineplot(data = flights, x = "month", y = "passengers")
plt.show()
sns.lineplot(data = flights, x = "month", y = "passengers", hue = "year")
plt.show()
```

```python
plt.subplot(2,2,1)
sns.lineplot(data = flights, x = "year", y = "passengers")
plt.subplot(2,2,2)
sns.boxplot(data = flights, x = "year", y = "passengers")
plt.show()
```

```python
plt.subplot(2,2,1)
sns.lineplot(data = flights, x = "month", y = "passengers")
plt.subplot(2,2,2)
sns.boxplot(data = flights, x = "month", y = "passengers")
plt.show()
```

#### pivot()함수로 테이블 재구성

```python
print(flights)
pvt = flights.pivot("month","year","passengers")
print(pvt)
```

#### 히트맵 그래프

```python
plt.figure(figsize=(10,6))
sns.heatmap(pvt, annot=True, fmt="d", cmap="YlGnBu")
plt.show()
```
